# **Student Performance Analysis**

### **STEP 1: Load Dataset**

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\jprat\Downloads\student\student-mat.csv", sep=';')
df.head()
```

### **STEP 2: Explore & Clean Data**

**Aim:** To check missing values.

```python
df.isnull().sum()
```

**Aim:** To Remove duplicates.

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

**Aim:** To Inspect dataset shape and data types.

```python
df.shape
df.dtypes
df.info()
```

### **STEP 3: Analysis Questions**

**Aim:** To determine the average final grade (G3) of students.

```python
df['G3'].mean()
```

**Result:** The average final grade (G3) is 10.42, indicating moderate student performance.

**Aim:** To determine the number of students who scored above 15 in the final grade (G3).

```python
len(df[df['G3'] > 15])
```

**Result:** A total of 40 students scored above 15, showing that high performance is achieved by a small group.

**Aim:** To examine the relationship between study time and student performance (G3).

```python
df['studytime'].corr(df['G3'])
```

**Result:** The correlation is 0.098, indicating a very weak positive relationship between study time and performance.

**Aim:** To compare the average final grades (G3) of male and female students.

```python
df.groupby('sex')['G3'].mean()
```

**Result:** Male students have an average grade of 10.91, while female students have 9.97, indicating male students perform slightly better, though the difference is small.

### **STEP 4: Visualizations**

**Aim:** To visualize the distribution of final grades (G3) of students using a histogram.

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

**Result** The histogram shows that most students scored between 8 and 14, with many students around 10 to 12.This means that most students have average performance. Only a few students scored very low or very high marks.

**Aim:** To understand the relationship between study time and final grades (G3) using a scatter plot.

```python
plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("G3")
plt.show()
```

**Result:** The scatter plot shows that students with higher study time tend to have slightly better grades.However, the points are spread out, which means the relationship is weak. 

**Aim:** To compare the average final grades (G3) of male and female students using a bar chart.

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

**Result:** The bar chart shows that male students have a slightly higher average grade compared to female students.

**Overall Conclusion:**

This suggests that study time alone does not strongly influence student performance.
